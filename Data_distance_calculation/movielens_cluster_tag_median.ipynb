```python
import pickle
from statistics import median
```

```python
with open('C:\\Users\\Java\\分群\\ml_cluster_tag_medians\\md2l.pkl' , 'rb') as f:
    comments = pickle.load(f)
comments[0]
```

```python
#group裡包含各個cluster的電影

groups = []  

for j in range(0,15):
    group = []
    for i in range(len(comments)):
        if comments[i]['label'] == j:
            group.append(comments[i])
    groups.append(group)

len(groups[1])
```

```python
# median_group1 = [] 第一個元素的中位數list

#每個group的各部電影的rel length都是一樣長的(1128)

#最後group的中位數list的length會有1128個, 也就是要先創造出1128個list

group_median_list = []

group_median_d = {} 

group_median = []
for n in range(len(groups)):                          #enter clusters
    medians = []
    for i in range(1128):                             #iteration times == element numbers        
        for_median = []
        for j in range(len(groups[n])):               #the numbers of movies in a cluster == all the elements required for one median
            for_median.append(groups[n][j]['rel'][i])
        medians.append(median(for_median))
    group_median.append(medians)

for i in range(15):
    group_median_d[str(i)] = group_median[i]
len(group_median)
with open('C:\\Users\\Java\\分群\\group_median', 'wb') as f:
    pickle.dump(group_median_d, f)
```
