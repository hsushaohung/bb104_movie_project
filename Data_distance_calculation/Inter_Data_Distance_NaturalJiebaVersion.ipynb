```python
documents_old = []
for i in range(len(comments)):
    single_movie = ""
    single_movie += comments[i]['words_jieba_stop']
    documents_old.append(single_movie)
documents_old[5:6] 
```

```python
vectorizer = TfidfVectorizer(stop_words = None)
X = vectorizer.fit_transform(documents_old)
 
# true_k = 15
# model = KMeans(n_clusters=true_k, init='k-means++', max_iter=300, n_init=15)
# model.fit(X)
```

```python
XX = X.toarray()
df = pd.DataFrame(XX)
df
```

```python
# df.values[0]
# ('aaa', df.values[0])

movie_vector_list = []

for i in range(len(comments)):
    for j in range(len(df)):
        if i == j:
            movie_vector_list.append((comments[i]['chname'], df.values[i]))
            break
movie_vector_list                                    
```

```python
#calculate inter-movie distances
#存成一個 List [chname, chname, distance]
inter_distances = [] 

for i in range(len(movie_vector_list)):
    distance_list = []
    for j in range(len(movie_vector_list)):               
        distance = numpy.sqrt(numpy.sum((movie_vector_list[i][1] - movie_vector_list[j][1])**2))
        distance_list.append([movie_vector_list[i][0], movie_vector_list[j][0], distance])
    inter_distances.append(distance_list)
    
#     big_list = sorted([small_list for small_list in movie_pair_list], key=lambda small_list:small_list[2], reverse=True)
#     print(big_list)
```

```python
inter_distances[0]
```

```python
#按照距離由近到遠排序 (由小到大)
sorted_distances = []

for i in range(len(inter_distances)):
    wan = inter_distances[i]
    wan = sorted(wan, key = lambda wan:wan[2])
    sorted_distances.append(wan)
```

```python
sorted_distances[0]
```

```python
recommend_dict_list = []
recommend_dict = {}
for i in range(len(sorted_distances)):
    recommend_dict['chname'] = sorted_distances[i][0][0]
    closest_movies = []
    for j in range(len(sorted_distances[i])):
        closest_movies.append(sorted_distances[i][j][1])
    recommend_dict['closest_movies'] = closest_movies
    recommend_dict_list.append({
                                         'chname': recommend_dict['chname'],
                                'closest_movies' : recommend_dict['closest_movies']
    })
```

```python
with open('C:\\Users\\Java\\分群\\Kmeans\\movie_distances.json', 'w', encoding = 'utf-8') as f:
    f.write(json.dumps(recommend_dict_list, ensure_ascii = False))
```
