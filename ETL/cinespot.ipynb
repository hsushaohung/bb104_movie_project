```python
import requests
from bs4 import BeautifulSoup
import pickle
import re
import json

DOMAIN = "http://www.cinespot.com"

URL = 'http://www.cinespot.com/hkmreviews'

resp = requests.get(URL)
soup = BeautifulSoup(resp.text, 'lxml')
                    
movie_dict_list = []
movieUrls = []

titles_num = soup.select('div.inner > p > a')
len(titles_num)

count = 0

for i in range(0, len(titles_num)):
    url_part = soup.select('div.inner > p > a')[i]['href']
    if 'hkmreviews' in url_part:
        critics_url = DOMAIN + url_part
        movieUrls.append(critics_url)                     #至此抓到所有網頁連結
        count+=1
        print(count)

with open("C:\\Users\\Java\\ETL\\movie\\Files_for_git\\url","wb") as f:
    pickle.dump(movieUrls,f)                              #先存成byte檔
```

```python
import requests
from bs4 import BeautifulSoup
import pickle
import re
import json
```

```python
with open("C:\\Users\\Java\\ETL\\movie\\Files_for_git\\url", "rb") as f:
    critics_link_urls = pickle.load(f) 

counter = 0    
movie_dict_list = []  

for critic_link in critics_link_urls:
    resp = requests.get(critic_link)
    soup = BeautifulSoup(resp.text, 'lxml')
    
    movie_dict = {}    
    
    if 'en' not in critic_link:                                                              #只抓中文影評
        splitter_list = [word.text for word in soup.select_one("div.inner").select("p")]     #先確定資訊頁和內容頁分隔在哪一行
        count = 1
        for word in splitter_list:
            if "主演" in word or "演員" in word:
                splitter = count
                break
            count+=1
        info = "".join(splitter_list[0:splitter])   #找出內文前的資訊

        movie_dict['chname']   = info.split('\n')[0]            
        movie_dict['comment'] = ["".join(splitter_list[splitter+1:])]                   #找出評論內文同時加進字典
        
                                                       
        movie_dict_list.append({ 
                                   'chname'    : movie_dict['chname'],
                                   'comment'   : movie_dict['comment']
            })
        counter+=1
        print(counter)
# movie_dict_list        

# with open('C:\\Users\\Java\\ETL\\movie\\Files_for_git\\cinespot.json', 'w', encoding = 'UTF-8') as f:
#     f.write(json.dumps(movie_dict_list, ensure_ascii = False))
```

```python
# 下次開起cinespot.json並且比對190部電影, 只留下需要的
with open('C:\\Users\\Java\\ETL\\movie\\InJ_compare','rb') as f:
    InJ_compare = pickle.load(f)
    
print(InJ_compare)
```

```python
upload_list = []

for m in movie_dict_list:
    for i in InJ_compare:
        if m['chname'] == i[3]:
            m['imdb_id'] = i[0]
            upload_list.append(m)
            break
upload_list
```
