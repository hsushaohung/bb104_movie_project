```python
import requests
from bs4 import BeautifulSoup as bs
import re
import json
import time
```

```python
DOMAIN = 'http://www.kingmedia.com.tw/'

URL_1 = DOMAIN + "cgi-bin/showhead.cgi?"

resp = requests.get(URL_1)
number_list = re.findall('"UID":"(\d+)"',resp.text)

name_list = []
columnists = resp.text.split('[{"')[1].split('"}]\r\n\n')[0].split('"},{"')
for i in range(0, len(columnists)):
    n = resp.text.split('[{"')[1].split('"}]\r\n\n')[0].split('"},{"')[i].split('"Nickname":"')[1]
    name_list.append(n)                 
    
print(name_list)
print(number_list)
```

```python
column_url = []

for number in number_list:
    URL_2 = DOMAIN + "cgi-bin/showBAlist.cgi?UID=%s"%number             #先抓每個影評的UID
    for page in range(0,90):
        URL_3 = URL_2 + "&p=%s"%page    
        column_url.append(URL_3)
print(column_url)    
```

```python
Success_url = []

for u in column_url:
    time.sleep(0.5)
    resp = requests.get(u)    
    if 'Success' not in resp.text:
        Success_url.append(u)
        
print(Success_url)
```

```python
Success_url[0]
```

```python
critic_urls = []

for s in Success_url:
    time.sleep(0.5)
    resp = requests.get(s)
    columns = json.loads(resp.text)
    critic_urls.append(DOMAIN + "cgi-bin/showarticle.cgi?Title={}&UID={}".format(columns[0]['Title'], columns[0]['UID']))

print(critic_urls)    
```

```python
import time
Titles = []

movie_dict_list = []
movie_dict = {}

for URL_4 in critic_urls:
    time.sleep(0.5)
    resp = requests.get(URL_4)    
    critic = json.loads(resp.text) 
    
    if '《' in critic[0]['Title']:
        movie_dict['chname']  = critic[0]['Title'].split('《')[1].split('》')[0]
    else: 
        movie_dict['chname']  = critic[0]['Title'].split(' /')[0]
    movie_dict['comment'] = [critic[0]['Message']]
    
    movie_dict_list.append({
                            'chname' : movie_dict['chname'],
                            'comment': movie_dict['comment']
    })
    
movie_dict_list

# 從190部裡面比對出吻合的標題並抓出他們的'Message'        
```
