```python
#HKfilmcriticssociety片名

import requests
from bs4 import BeautifulSoup
import pickle
import re
import json

DOMAIN = "http://www.filmcritics.org.hk"

headers = {
    "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.84 Safari/537.36"
}

resp = requests.get(DOMAIN + "/taxonomy/term/1?page=0", headers = headers)
resp.text

movie_dict_list = []

pageurl_list = []

lastpg = int(re.findall('href="(.+)" title="到最後一頁"', resp.text)[0].split('page=')[1])  #給出最後一頁的頁數,找出總共有幾頁
for pg in range(0, lastpg +1):
    p = DOMAIN + "/taxonomy/term/1?page=%s"%pg
    pageurl_list.append(p)
# print(page_list) 

# with open('C:\\Users\\Java\\ETL\\movie\\hkfcsUrls','wb') as f:
#     pickle.dump(pageurl_list,f)
    
# with open('C:\\Users\\Java\\ETL\\movie\\hkfcsUrls','rb') as f:
#     page_list = pickle.load(f)

article_urls = []
movie_dict = {}
counter = 0

for page in pageurl_list:
    resp = requests.get(page, headers = headers)
    soup = BeautifulSoup(resp.text, 'lxml')
    for i in range(1,11):
        link = soup.select('div.views-row-%s > div.node > h2 > a'%i)[0].text
        article_urls.append(link)

movie_names = []
for article_url in article_urls:
    if '《' in article_url:
        movie_name = article_url.split('《')[1].split('》')[0]
        movie_names.append(movie_name)

movie_namess = list(set(movie_names))
print(movie_namess)
```

```python
#動映地帶 片名

import requests
from bs4 import BeautifulSoup
import pickle
import re
import json

DOMAIN = "http://www.cinespot.com"

URL = 'http://www.cinespot.com/hkmreviews'

resp = requests.get(URL)
soup = BeautifulSoup(resp.text, 'lxml')
                    
movie_dict_list = []

movieUrls = []
movie_dict = {} 
names = []

titles_num = soup.select('div.inner > p > a')
len(titles_num)

counter = 0

for i in range(0, len(titles_num)):
    url_part = soup.select('div.inner > p > a')[i]['href']
    if 'hkmreviews' in url_part:
        critics_url = DOMAIN + url_part
        movieUrls.append(critics_url) 
        
for critic_link in movieUrls:
    resp = requests.get(critic_link)
    soup = BeautifulSoup(resp.text, 'lxml')    
    if 'en' not in critic_link:                                                              #只抓中文影評
        splitter_list = [word.text for word in soup.select_one("div.inner").select("p")]     #先確定資訊頁和內容頁分隔在哪一行
        count = 1
        for word in splitter_list:
            if "主演" in word or "演員" in word:
                splitter = count
                break
            count+=1
        info = "".join(splitter_list[0:splitter])   #找出內文前的資訊
        name = info.split('\n')[0]
        names.append(name)
        counter+=1
        print(counter)
print(names)
```
